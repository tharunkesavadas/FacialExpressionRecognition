```
import tensorflow as tf
import keras
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

for _ in range(4):
  expression = [ "Happy", "Neutral", "Sad" ,"Surprise"] 
  image = cv2.imread(input("Enter your image path to recognise face expression:"))
  image = cv2.resize(image, (300,200))
  cv2_imshow(image)
  image = cv2.resize(image, (48,48))
  model = load_model("/content/drive/My Drive/epoch 40 latest/trainedmodel_epoch40.h5")
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  imageArray = img_to_array(image)
  imageArray = np.expand_dims(imageArray, axis = 0)
  y = model.predict(x = imageArray)
  l=list(y[0])
  print("Probabilities of predicted classes:",l)
  temp = l.index(max(l))
  print("Predicted Class Number:",temp)
  print("Predicted Expression:",expression[temp])
  print("_______________________________________"*3)
```
